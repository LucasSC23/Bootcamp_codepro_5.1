```python
#Escribe un programa que pregunte al usuario su edad para ingresar al festival Asunciónico y determine qué tipo de acceso puede tener.

#Si el usuario tiene menos de 18 imprimir: "Acceso denegado — Solo zona Kids 🎈"
#Si el usuario tiene mas de 18 pero menos de 21 imprimir: "Acceso permitido — No puede comprar bebidas alcoholicas 🎟️"
#Si el usuario tiene mas de 21 imprimir: "Acceso total — Disfruta el festival 🎉"


edad=int(input("Ingrese su edad: "))
if edad < 18:
    print("Acceso denegado — Solo zona Kids 🎈")
elif edad <= 21:
    print("Acceso permitido — No puede comprar bebidas alcoholicas 🎟️")
else:
    print("Acceso total — Disfruta el festival 🎉")
```

```python
#Escribir un programa en el que se pregunte al usuario por una temperatura 
# (en grados Celsius), imprima la temperatura escrita, y luego clasificar la 
# temperatura ingresada en los siguientes niveles:

#Entre -10 y 13 grados, que imprima "Ro'y koko".
#Entre 14 y 23 grados, que imprima "Refresca'i".
#Entre 24 y 28 grados, que imprima "Oikoite".
#Entre 29 y 45 grados, que imprima "Hakurooooo".
#Para todo lo demás, imprimir "Ñamano mba'e!".

temperatura_usuario=float(input("Ingrese la temperatura"))

def temp_impresion(temperatura):
    if temperatura >= -10 and temperatura <= 13 :
        print(f"La tempertura es: {temperatura}")
        print("Ro'y koko")
        
    elif temperatura >= 14 and temperatura <= 23 :
        print(f"La tempertura es: {temperatura}")
        print("Refresca'i")
        
    elif temperatura >= 24 and temperatura <= 28 :
        print(f"La tempertura es: {temperatura}")
        print("Oikoite")
        
    elif temperatura >= 29 and temperatura <= 45 :
        print(f"La tempertura es: {temperatura}")
        print("Hakurooooo")
        
    else:
        print(f"La tempertura es: {temperatura}")
        print("Ñamano mba'e!")
        
temp_impresion(temperatura_usuario)
```

```python
#Escribir un programa que almacene una contraseña en una variable, y pregunte
# al usuario por la contraseña hasta que introduzca la contraseña correcta.
#Cuando el usuario ingrese la contrasena correcta, imprimir un mensaje secreto.

contraseña= "Holamundo."
while True:
    contraseña_usuario=(input("Ingrese su contraseña"))
    if contraseña_usuario != contraseña:
        print("Contraseña incorrecta. Ingrese de nuevo")
    else:
        print("Ya salio del bucle, gracias por usar nuestro programa")
        break
```

```python
#Solicitar al usuario que ingrese una frase.
# Imprimir un mensaje cada vez que se encuentre una letra "o".
frase=(input("Ingrese una frase"))
contador= 0
for i in frase:
    if i =="o" or i =="O":
        print("Encontramos una o/O")
        contador+=1
print(f"Encontramos {contador}, letras o/O \nEn la frase {frase} ")

        
        
```
