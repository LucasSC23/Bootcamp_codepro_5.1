```python
class Coche: 
    tipo = "vehiculo" # Atributo de clase

    def __init__(self, modelo, posicion=0):
        self.modelo = modelo
        self.posicion = posicion

    def acelerar(self, metros):
        self.posicion += metros # Forma corta de decir metros + posicion
        print(f"{self.modelo} acelera y avanza {metros}m. Nueva posición: {self.posicion}")

    def girar(self, direccion):
        print(f"{self.modelo} gira a la {direccion}")

    def frenar(self): # Corregido el nombre (antes decía girar)
        print(f"{self.modelo} frena y se detiene en la posición {self.posicion}")

# --- AQUÍ EMPIEZA LA EJECUCIÓN (Fuera de la clase) ---
mi_coche = Coche("Ranger Raptor")
mi_coche.acelerar(50)
mi_coche.frenar()
mi_coche.girar("derecha")
```

Definir una clase madre que herede a sus clases hijas dos atributos de instancia. Mostrar ambos atributos en pantalla desde los objetos de las clases hijas.
