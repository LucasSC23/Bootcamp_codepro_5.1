```python
#Practicar variables globales y locales.Escribir una función calcular_total que reciba 
# el precio de un producto y calcule el total sumando un impuesto global impuesto = 10.
#Usa una variable local total para almacenar el resultado e imprime el total.
#Llama a la función con distintos precios para probar.

#variable
producto=int(input("Ingrese el valor del producto"))
impuesto=0.1
#Funcion
def calcular_total(precio):
    calcular_impuesto = precio+(precio*impuesto)
    print(f"El total es: {calcular_impuesto}")
#llamado de funcion    
calcular_total(producto)
```

```python
#Imagina que estás repartiendo caramelos: solicita al usuario ingresar  el total de caramelos y la cantidad 
# de niños, luego calcula cuántos caramelos recibe cada niño usando división entera y cuántos sobran usando módulo e imprime ambos resultados.

#Variables
cant_caramelos = int(input("Ingrese la cantidad de caramelos por favor: "))
cant_niños = int(input("Ingrese la cantidad de niños por favor: "))
#cálculo
cant_recibida= cant_caramelos / cant_niños
modulo= cant_caramelos % cant_niños
#Impresión
print(f"La cantidad de caramelos por niños es: {cant_recibida}")
print(f"y los caramelos que sobran son {modulo}")      
```
